```python
import re

# Let's fetch the data using curl (for now)
result = !curl -s "http://w4.stern.nyu.edu/faculty/search_name_form"

# The string.join(list of strings) 
# is an idiom of Python for takings all the strings in a list and connecting them
# using the string as the concetenating character
html = " ".join(result) 

# This is an attempt to write a regex for emails. 
# It is not perfect, but works for our purposes.
# Sidenote: the perfect regex for emails is at http://ex-parrot.com/~pdw/Mail-RFC822-Address.html
# Yep, it is 80+ lines :-)
email_regex = re.compile('[A-Za-z0-9\.]+@([A-Za-z0-9\.]+\.)+[A-Za-z]{2,3}')
matches = email_regex.finditer(html)

for m in matches:
    email = m.group().strip()
    print "Email:", email
```
